```python
# import libraries

import requests
from bs4 import BeautifulSoup
```

```python
# request a sample onepetro page
# key = "well test"

url = 'https://onepetro.org/search?q=well+test&peer_reviewed=&published_between=&from_year=&to_year=&rows=25'
r = requests.get(url)
```

```python
# print response

r.text
```

```python
# navigate html

soup = BeautifulSoup(r.text, 'html.parser')
```

```python
soup.title
```

```python
print(soup.find_all('div class="result-item"', limit=1)[0].prettify())
```

```python
print(soup.find_all('div', {'class':'result-item'})[0].get_text())
```

```python
print(soup.find_all('div', {'class':'result-item'})[0].prettify())
```
