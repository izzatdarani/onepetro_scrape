```python
# import lib

import requests
from bs4 import BeautifulSoup
from fake_useragent import UserAgent
import random
```

```python
# learning how to use ip proxy in requests

# create user agent
ua = UserAgent(cache=False)

# get proxy list from url below
url = 'https://www.sslproxies.org/'

# requesting the url to get list of ip proxies
headers = {'User-Agent':ua.random}
proxy_response = requests.get(url, headers=headers)
```

```python
# check response
proxy_response.status_code
```

```python
# text preview
proxy_response.text
```

```python
# soup the related content
soup = BeautifulSoup(proxy_response.text, 'html.parser')

proxies_table = soup.find(id='proxylisttable')

# Save proxies in the array
for i,row in enumerate(proxies_table.tbody.find_all('tr')):
    print(i, row.find_all('td')[0].string, row.find_all('td')[1].string)
```
